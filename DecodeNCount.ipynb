```python
import csv
import os

#用OS抓取系統分隔符號
sep = os.sep
#字典
tran = {
    "a" : "y",
    "b" : "z",
    "c" : "c",
    "d" : "a",
    "e" : "b",
    "f" : "d",
    "g" : "e",
    "h" : "h",
    "i" : "f",
    "j" : "g",
    "k" : "i",
    "l" : "j",
    "m" : "m",
    "n" : "k",
    "o" : "o",
    "p" : "p",
    "q" : "l",
    "r" : "n",
    "s" : "s",
    "t" : "t",
    "u" : "r",
    "v" : "q",
    "w" : "h",
    "x" : "u",
    "y" : "v",
    "z" : "x"
}

#解密
def decod(url):
    buf = ""
    for i in url:
        if i.isalpha():
            buf+=tran[i]
        else:
            buf+=i
    return buf

cust = [] #客戶ID
act = [] #客戶行為，Index與ID相對
#是否為第一行
firstline = True
lcnt = 0
#檔案路徑
pat = "TBN_CUST_BEHAVIOR.csv"
#讀取檔案
with open(pat, newline='') as csvfile:
    print("Reading File...")
    rows = csv.reader(csvfile)
    for row in rows:
        lcnt+=1
        print(str(lcnt),end="\r")
        if firstline:    #skip first line
            firstline = False
            continue
        else:
            if row[0] not in cust:
                cust.append(row[0])
                #陣列中四個值分別為：
                #CreditCard,Forex,Trust,Loan
                act.append([0,0,0,0])
            idx = cust.index(row[0])
            rep = row[2].replace("http://www.esunbank.com.tw/","").replace("https://www.esunbank.com.tw/","")
            #讀取以後先解密
            url = decod(rep)
            if "credit-card" in url:
                act[idx][0]+=1
            elif "forex" in url:
                act[idx][1]+=1
            elif "trust" in url:
                act[idx][2]+=1
            elif "loan" in url:
                act[idx][3]+=1
                
#寫入結果，選擇性
#可以改成把陣列內容直接抓出來
f = open("res.txt", "a")
f.write("ID,CreditCard,ForEx,Trust,Loan\n")
for cid in cust:
    idx = cust.index(cid)
    buf = cid," - CreditCard: ",str(act[idx][0]),", "," - ForEx: ",str(act[idx][1]),", "," - Trust: ",str(act[idx][2]),", "," - Loan: ",str(act[idx][3])
    f.write(cid+","+str(act[idx][0])+","+str(act[idx][1])+","+str(act[idx][2])+","+str(act[idx][3])+"\n")
    print(buf)
f.close()
```
