```python
import pandas as pd
import numpy as np
# import tensorflow as tf
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
```

```python

df = pd.read_csv('cif_combine_browse.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
#挑出 答案卷 30,000 筆的資料
answer=pd.read_csv('TBN_Y_ZERO.csv')
ans_arr=[] # 在答案卷裡面的 id

col=[]

for i in answer.iloc[:,0]:
    ans_arr.append(i)
    
x_test_index=[] # 把答案卷裡面的 (predict_index)
x_train_index = [] # 準備用來training 的 array (train_index)

for index,row in df.iterrows():
    if row.values[0] in ans_arr:
        x_test_index.append(index)
    else :
        x_train_index.append(index)
```

```python
df = df.drop(['GENDER_CODE'], axis=1)
```

```python
df.head()
```

```python
def checking_loss(results):
    plt.figure(figsize=(8, 8))
    plt.title("Learning curve")
    plt.plot(results.history["loss"], label="loss")
    plt.plot(results.history["val_loss"], label="val_loss")
    plt.plot( np.argmin(results.history["val_loss"]), np.min(results.history["val_loss"]), marker="x", color="r", label="best model")
    plt.xlabel("Epochs")
    plt.ylabel("log_loss")
    plt.legend();
```

```python
#切成training data 和 testing data

df_x_test= df.iloc[x_test_index,:]
df_x_train=df.iloc[x_train_index,:]
```

```python
df_x_test.shape
```

```python
df_x_train.shape
```

```python
# 分別刪除各項金融產品建立不同的model

CreditCard = ['CUST_NO','ForEx_b','fx_BuyFreq','Trust_b','ln_BuyFreq','Loan_b','wm_BuyFreq']

ForEx =['CUST_NO','CreditCard_b','cc_BuyFreq','Trust_b','ln_BuyFreq','Loan_b','wm_BuyFreq']

Loan = ['CUST_NO','CreditCard_b','cc_BuyFreq','ForEx_b','fx_BuyFreq','Trust_b','wm_BuyFreq']

Trust = ['CUST_NO','CreditCard_b','cc_BuyFreq','ForEx_b','fx_BuyFreq','Loan_b','ln_BuyFreq']

# CreditCard = ['CUST_NO','ForEx_b','fx_BuyFreq','Trust_b','ln_BuyFreq','Loan_b','wm_BuyFreq',
#               'AGE','CHILDREN_CNT','EDU_CODE','INCOME_RANGE_CODE','WORK_MTHS']

# ForEx = ['CUST_NO','CreditCard_b','cc_BuyFreq','Trust_b','ln_BuyFreq','Loan_b','wm_BuyFreq',
#          'AGE','CHILDREN_CNT','EDU_CODE','INCOME_RANGE_CODE','WORK_MTHS']

# Loan = ['CUST_NO','CreditCard_b','cc_BuyFreq','ForEx_b','fx_BuyFreq','Trust_b','wm_BuyFreq',
#        'AGE','CHILDREN_CNT','EDU_CODE','INCOME_RANGE_CODE','WORK_MTHS']

# Trust = ['CUST_NO','CreditCard_b','cc_BuyFreq','ForEx_b','fx_BuyFreq','Loan_b','ln_BuyFreq',
#         'AGE','CHILDREN_CNT','EDU_CODE','INCOME_RANGE_CODE','WORK_MTHS']

# training data
df_CreditCard_train = df_x_train.drop(CreditCard, axis=1)
df_ForEx_train = df_x_train.drop(ForEx, axis=1)
df_Loan_train = df_x_train.drop(Loan, axis=1)
df_Trust_train = df_x_train.drop(Trust, axis=1)

# testing data
df_CreditCard_test = df_x_test.drop(CreditCard, axis=1)
df_ForEx_test = df_x_test.drop(ForEx, axis=1)
df_Loan_test = df_x_test.drop(Loan, axis=1)
df_Trust_test = df_x_test.drop(Trust, axis=1)
```

```python
#信用卡
#機器學習 Machine Learning
#切分訓練與驗證資料 Separate data in training and validation set

#把label切分出來
x_train_CreditCard=df_CreditCard_train.drop('cc_BuyFreq',axis=1)
x_test_CreditCard=df_CreditCard_test.drop('cc_BuyFreq',axis=1)
y_train_CreditCard=df_CreditCard_train['cc_BuyFreq']
y_test_CreditCard=df_CreditCard_test['cc_BuyFreq']
```

```python
# Feature scaling 特徵工程
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train_CreditCard)
X_test = sc.fit_transform(x_test_CreditCard)
```

```python
X_train.shape[1]
```

```python
y_train_CreditCard.head()
```

```python
y_train_CreditCard.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from keras.callbacks import ReduceLROnPlateau
from keras.callbacks import ModelCheckpoint
from sklearn.model_selection import cross_val_score
from keras import backend as K
from keras.layers import BatchNormalization
```

```python
# 設定checkpoint & callbacks
def callbacks(name):
    save_dir = "checkpoint"
    #     tl.files.exists_or_mkdir(save_dir)
    path=save_dir+'/model_check_'+name+'.h5'
    callbacks = [
        EarlyStopping(patience=10, verbose=1),
        ReduceLROnPlateau(factor=0.1, patience=10, min_lr=0.000001, verbose=1),
        ModelCheckpoint(path, verbose=1, save_best_only=True, save_weights_only=False)
    ]
    callbacks=None
    return callbacks
```

```python
seed = 12 
def create_model_cc():
    model = Sequential()
    model.add(Dense(20, input_dim=X_train.shape[1],init='uniform', activation='relu'))
    #model.add(BatchNormalization())
    #model.add(Dropout(0.5))
    #model.add(Dropout(0.5))
    model.add(Dense(10, init='uniform', activation='relu'))
    model.add(Dense(10, init='uniform', activation='relu'))
    model.add(Dense(1, activation='relu'))
    model.compile(loss='binary_crossentropy' , optimizer='adam', metrics=['accuracy'])  
    model.summary()
    return model
```

```python
model_cc = create_model_cc()
history = model_cc.fit(X_train, y_train_CreditCard, validation_split=0.20, 
                       epochs=200, batch_size=200, verbose=2,shuffle=True,callbacks=callbacks('cc'))
```

```python
import matplotlib.pyplot as plt

# plot the history of the training model
def history_plot(history): 
    # list all data in history
    print(history.history.keys())
    # summarize history for accuracy
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
history_plot(history)
```

```python
checking_loss(history)
```

```python
# Print Accuracy
scores = model_cc.evaluate(X_test, y_test_CreditCard) 
print("accuracy=",scores[1])
```

```python
#predict
predicion_CreditCard_rf=model_cc.predict_classes(X_test)

```
